```python
# SMTP stand for Simpler Mail Transfer Protocol
import smtplib
```

```python
# 587 is a port number (TLS encyption)
smtp_object = smtplib.SMTP('smtp.gmail.com', 587)
```

```python
# tell that your connection is success
smtp_object.ehlo()
```

```python
smtp_object.starttls()
```

```python
password = input('What is your password: ')
```

```python
# make input password invisible
import getpass
```

```python
password = getpass.getpass('Password: ')
```

```python
# password from gmail app password
email = getpass.getpass('Email: ')
password = getpass.getpass('Password: ')
smtp_object.login(email, password)
```

```python
from_address = email
to_address = email
subject = input('enter the subject line: ')
message = input('enter the body message: ')
msg = 'Subject: '+subject+'\n'+message

smtp_object.sendmail(from_address, to_address, msg)
```

```python
# close connection (will get error if you did not connect first)
smtp_object.quit()
```

```python
 import imaplib
```

```python
M = imaplib.IMAP4_SSL('imap.gmail.com')
```

```python
import getpass
```

```python
email = getpass.getpass('Email: ')
password = getpass.getpass('Password: ')
```

```python
M.login(email, password)
```

```python
M.list()
```

```python
M.select('inbox')
```

```python
typ, data = M.search(None, 'SUBJECT "Test Sending Email with Python"')
```

```python
typ
```

```python
data
```

```python
email_id = data[0]
```

```python
result, email_data = M.fetch(email_id, '(RFC822)')
```

```python
email_data
```

```python
raw_email = email_data[0][1]
```

```python
raw_email_string = raw_email.decode('utf-8')
```

```python
import email
```

```python
email_message = email.message_from_string(raw_email_string)
```

```python
for part in email_message.walk():
    
    if part.get_content_type() == 'text/plain':
        body = part.get_payload(decode= True)
        print(body)
```
