```python
text = 'my phone once, my phone twice'
```

```python
# return position of where firt match founded
match = re.search('phone', text)
```

```python
match
```

```python
# find all match
matches = re.findall('phone', text)
```

```python
matches
```

```python
len(matches)
```

```python
# find position of all match founded
for match in re.finditer('phone', text):
    print(match.span())
    
```

```python
for match in re.finditer('phone', text):
    print(match.group())
```

```python
# Find pattern using character identifiers
```

```python
text = 'My phone number is 082-499-7130'
```

```python
phone = re.search(r'\d\d\d-\d\d\d-\d\d\d\d',text)
```

```python
phone
```

```python
phone.group()
```

```python
phone = re.search(r'\d{3}-\d{3}-\d{4}',text)
```

```python
phone
```

```python
phone_pattern = re.compile(r"(\d{3})-(\d{3})-(\d{4})")
```

```python
results = re.search(phone_pattern, text)
```

```python
results.group()
```

```python
results.group(1)
```

```python
results.group(2)
```

```python
results.group(3)
```

```python
results.group(4)
```
